```python
import pandas as pd
from openpyxl import load_workbook
import os

# Path to the desktop on your cloud drive
desktop_path = '/Users/hannahkaufman/Desktop'

def process_excel(file_path, sheet_name='Sheet0', participant1_num=1, participant2_num=2, combined_df=None):
    """
    Processes the Excel file to extract data items and categories, and updates the combined DataFrame.

    Parameters:
    - file_path: str, path to the Excel file.
    - sheet_name: str, name of the sheet to process.
    - participant1_num: int, participant number for the first participant.
    - participant2_num: int, participant number for the second participant.
    - combined_df: DataFrame, the combined DataFrame to which data will be appended.

    Returns:
    - combined_df: DataFrame, the updated combined DataFrame with data from the processed file.
    """
    # Load the workbook and select the specified sheet
    workbook = load_workbook(filename=file_path)
    sheet = workbook[sheet_name]

    # Define the column range from W to GH (columns 23 to 190)
    start_col = 23  # Column W
    end_col = 190  # Column GH

    # Initialize lists to store the data items for each participant
    data_item_array1 = []
    data_item_array2 = []

    # Iterate through the cells in row 2 from column W to column GH to find data items and categories
    for col in range(start_col, end_col + 1):
        cell_value = sheet.cell(row=2, column=col).value
        if cell_value and "- Health Data Category - " in cell_value:
            parts = cell_value.split("- Health Data Category - ")
            if len(parts) > 1:
                # Extract the data item and associated categories for both participants
                data_item = parts[1].strip()
                category1 = sheet.cell(row=3, column=col).value
                data_items1 = [data_item, category1]  # Remove column number
                data_item_array1.append(data_items1)
                category2 = sheet.cell(row=4, column=col).value
                data_items2 = [data_item, category2]  # Remove column number
                data_item_array2.append(data_items2)

    # Create DataFrames for participant 1 and participant 2 without the column number
    df1 = pd.DataFrame(data_item_array1, columns=['Data Item', 'Categories'])
    df1['Participant ID'] = f'Participant {participant1_num}'  # Add Participant ID column

    df2 = pd.DataFrame(data_item_array2, columns=['Data Item', 'Categories'])
    df2['Participant ID'] = f'Participant {participant2_num}'  # Add Participant ID column

    # Concatenate the two DataFrames row-wise to form the processed DataFrame
    processed_df = pd.concat([df1, df2], axis=0)
    processed_df.reset_index(inplace=True, drop=True)

    # Add additional columns with default values (0) to the DataFrame for each health category
    additional_columns = ["Behavioral health", "Diagnoses", "Disabilities", "Infectious diseases",
                          "Genetics", "Medications", "Sexual and reproductive health",
                          "Social determinants of health", "Other"]

    for col in additional_columns:
        processed_df[col] = 0

    # Iterate over each row in the processed DataFrame to update category columns
    for index, row in processed_df.iterrows():
        # Split the categories in the "Categories" column by comma and strip leading/trailing spaces
        for category in row['Categories'].split(','):
            category = category.strip()
            if category in processed_df.columns:
                # Update the corresponding column in that row to 1 if the category is present
                processed_df.at[index, category] = 1

    # Print the processed DataFrame for verification
    print(processed_df)

    # Reorder columns to place "Participant ID" first
    cols = ['Participant ID', 'Data Item'] + additional_columns
    processed_df = processed_df[cols + ['Categories']]  # Ensure 'Categories' column is last

    # Append the processed DataFrame to the combined DataFrame
    if combined_df is not None:
        combined_df = pd.concat([combined_df, processed_df], ignore_index=True)
    else:
        combined_df = processed_df

    return combined_df

def process_all_files(file_paths, output_dir, sheet_name='Sheet0'):
    """
    Processes all the specified Excel files and combines their data into a single DataFrame.

    Parameters:
    - file_paths: list of str, paths to the Excel files.
    - output_dir: str, directory to save the combined output file.
    - sheet_name: str, name of the sheet to process in each file.
    """
    combined_df = None  # Initialize an empty DataFrame to hold combined data

    # Iterate over the file paths and process each file
    for i, file_path in enumerate(file_paths):
        participant1_num = i * 2 + 1  # Calculate participant 1 number based on file index
        participant2_num = i * 2 + 2  # Calculate participant 2 number based on file index
        # Process the current Excel file and update the combined DataFrame
        combined_df = process_excel(file_path, sheet_name=sheet_name, participant1_num=participant1_num, participant2_num=participant2_num, combined_df=combined_df)
        print(f"Processed file {file_path} with participants {participant1_num} and {participant2_num}")

    # Save the combined DataFrame to an Excel file on the desktop
    output_file_path = os.path.join(output_dir, 'Combined_Participants_Data.xlsx')
    combined_df.to_excel(output_file_path, index=False)
    print(f"Combined DataFrame has been saved to {output_file_path}")

# Example usage
file_paths = [
    '/Users/hannahkaufman/Dropbox (ASU)/Physicians Views on Sensitive Data/Results & Agreements/Patient 1.xlsm',
    '/Users/hannahkaufman/Dropbox (ASU)/Physicians Views on Sensitive Data/Results & Agreements/Patient 2.xlsm',
    '/Users/hannahkaufman/Dropbox (ASU)/Physicians Views on Sensitive Data/Results & Agreements/Patient 3.xlsm',
    '/Users/hannahkaufman/Dropbox (ASU)/Physicians Views on Sensitive Data/Results & Agreements/Patient 4.xlsm',
    '/Users/hannahkaufman/Dropbox (ASU)/Physicians Views on Sensitive Data/Results & Agreements/Patient 5.xlsm',
    '/Users/hannahkaufman/Dropbox (ASU)/Physicians Views on Sensitive Data/Results & Agreements/Patient 6.xlsm',
    '/Users/hannahkaufman/Dropbox (ASU)/Physicians Views on Sensitive Data/Results & Agreements/Patient 7.xlsm',
    '/Users/hannahkaufman/Dropbox (ASU)/Physicians Views on Sensitive Data/Results & Agreements/Patient 8.xlsm',
    '/Users/hannahkaufman/Dropbox (ASU)/Physicians Views on Sensitive Data/Results & Agreements/Patient 9.xlsm',
    '/Users/hannahkaufman/Dropbox (ASU)/Physicians Views on Sensitive Data/Results & Agreements/Patient 10.xlsm',
    '/Users/hannahkaufman/Dropbox (ASU)/Physicians Views on Sensitive Data/Results & Agreements/Patient 11.xlsm',
    '/Users/hannahkaufman/Dropbox (ASU)/Physicians Views on Sensitive Data/Results & Agreements/Patient 12.xlsm'
]

# Output directory is the desktop path
output_dir = desktop_path

# Process all files and combine their data into a single DataFrame
process_all_files(file_paths, output_dir, sheet_name='Sheet0')
```
